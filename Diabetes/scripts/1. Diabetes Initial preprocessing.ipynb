```python
# Function to read and merge csv files in a folder to a dataframe
def merge_csv_files_in_folder(folder_path):
    # Create an empty list to store DataFrames
    dfs = []

    # Define the expected prefix of the filenames
    filename_prefix = "DiabetesAtlas_CountyData_"

    # Iterate through the files in the folder
    for filename in os.listdir(folder_path):
        if filename.endswith(".csv") and filename.startswith(filename_prefix):
            # Attempt to extract the year from the filename
            try:
                year = int(filename[len(filename_prefix):-4])
            except ValueError:
                # If the year extraction fails, skip the file
                print(f"Skipping file {filename} due to invalid year format.")
                continue
            
            file_path = os.path.join(folder_path, filename)
            
            # Determine the number of lines in the file
            with open(file_path) as f:
                num_lines = sum(1 for _ in f)
            
            # Read the CSV file, skipping the first two rows and the last row
            df = pd.read_csv(file_path, skiprows=2, skipfooter=1, engine='python')
            
            # Add a "Year" column with the extracted year
            df["Year"] = year
            
            # Append the DataFrame to the list
            dfs.append(df)

    # Concatenate all DataFrames into a single DataFrame
    if dfs:
        final_df = pd.concat(dfs, ignore_index=True)
        # print("Concatenation successful.")
        return final_df
    else:
        print("No CSV files matching the expected format found in the folder.")
        return None
```
