#### Importing Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm

# code to ignore warnings
import warnings
warnings.filterwarnings("ignore")
```

#### Reading Files

```python
file_path = "Diab_County_Data_New.xlsx"
df_diab = pd.read_excel(file_path)
```

```python
# Restricting to required columns
df_diab = df_diab[['State','County','CountyFIPS','Year','Tax_Type','Rate','Total Absolute','Total Percentage','Gender','Gender Absolute','Gender Percentage','Age','Age Diabetics Absolute','Age Diabetics Percentage','Notes_On_Tax_Rate']]
```

```python
# Check for data type
df_diab.dtypes
```

```python
# Converting data type
columns_to_convert = ['Total Absolute', 'Total Percentage', 'Gender Absolute', 'Gender Percentage', 'Age Diabetics Absolute', 'Age Diabetics Percentage']
for column in columns_to_convert:
    df_diab[column] = pd.to_numeric(df_diab[column], errors='coerce', downcast='float')
```

```python
df_diab
```

```python
# Check for Null Values
df_diab.isna().sum()
```

```python
df_diab.dropna(subset={'Total Percentage','Total Absolute'},inplace=True)
```

```python
# Pivot the Tax_Type column
df_diab_pivot = df_diab.pivot_table(index=['State', 'County', 'CountyFIPS', 'Year', 'Total Absolute','Total Percentage',
                                      'Gender', 'Gender Absolute', 'Gender Percentage','Age', 'Age Diabetics Absolute',
                                     'Age Diabetics Percentage'],
                           columns='Tax_Type', values='Rate', aggfunc='mean').reset_index()
# Rename the columns for clarity
df_diab_pivot.columns.name = None  # Remove the 'Tax_Type' label

df_diab_pivot
```

```python
df_diab_pivot.isna().sum()
```

#### Exploratory Data Analysis

Before moving to Scatter plot and Pearson Correlation, lets focus only on total numbers and percentage and remove duplicates established due to gender and age:

```python
df_diab_total_pivot = df_diab_pivot[['State','County','CountyFIPS', 'Year', 'Total Absolute','Total Percentage','Grocery Tax','Restaurant Tax']]
df_diab_total_pivot.drop_duplicates(inplace=True)
```

```python
df_diab_total_pivot
```

##### 1. Grocery Tax

a. Total Absolute Diabetics vs Grocery Tax

```python
plt.scatter(df_diab_total_pivot['Grocery Tax'], df_diab_total_pivot['Total Absolute'], marker='o', color='blue', alpha=0.7)

# Add labels and title
plt.xlabel('Grocery Tax Rate')
plt.ylabel('Total Absolute')
plt.title('Scatter Plot of Total Absolute Diabetics and Grocery Tax')
```

##### Conclusion:
At higher grocery tax rates, the total absolute number is concentrated to lower bandwidth. However, this is not a causation. It is possible that since we are looking at totals, the counties/states with high grocery tax rate have less population and hence less absolute number in diabetics. Hence we will draw a plot on Percentage

b. Total Percentage Diabetics vs Grocery Tax

```python
plt.scatter(df_diab_total_pivot['Grocery Tax'], df_diab_total_pivot['Total Percentage'], marker='o', color='blue', alpha=0.7)

# Add labels and title
plt.xlabel('Grocery Tax Rate')
plt.ylabel('Total Percentage')
plt.title('Scatter Plot of Total Percentage Diabetics and Grocery Tax')
```

#### Conclusion:
This plot helps us quash the possible causation based on absolute number. Looks like the percentage diabetics is not much influenced by grocery tax. Lets do a Pearson correlation test to understand more.

c. Pearson Correlation of Total Absolute with Grocery Tax

```python
correlation = df_diab_total_pivot['Grocery Tax'].corr(df_diab_total_pivot['Total Absolute'])
print('Grocery Tax - Total Absolute ',correlation)
```

#### Conclusion:
This correlation shows that total absolute has negative correlation with grocery tax i.e. as grocery tax increases the total absolute number decreases. However, this is again on absolute numbers and does not account for population number. Hence we will do a Pearson Correlation on Percentage.

d. Pearson Correlation of Total Percentage with Grocery Tax

```python
correlation = df_diab_total_pivot['Grocery Tax'].corr(df_diab_total_pivot['Total Percentage'])
print('Grocery Tax - Total Percentage',correlation)
```

#### Conclusion:
This correlation shows that total percentage has a minor positive correlation with grocery tax i.e. as grocery tax increases the total percentage of diabetics patient increases.

#### Regression Analysis at County level data

1. Baseline without controls

a. With Total Absolute as Response Variable:

##### Note: 
Even though Total Absolute does not make sense as concluded from EDA, we are carrying regression on it.

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Assuming 'Grocery Tax' and 'Restaurant Tax' are the features
X = df_diab_total_pivot[['Grocery Tax', 'Restaurant Tax']]
y = df_diab_total_pivot['Total Absolute']

# Add a constant term to the features matrix (required for statsmodels)
X = sm.add_constant(X)

# Create and fit the linear regression model using statsmodels
stats_model = sm.OLS(y, X).fit()

# Print the summary
print(stats_model.summary())
```

#### Conclusion:
The baseline model with response as total number of diabetics and predictor variables grocery tax and restaurant tax:
- One percent increase in grocery tax, keeping all other constant, decreases the absolute number by 9.943 x 10^4
- One percent increase in restaurant tax, keeping all other constant, increases the absolute number by 1.064 x 10^5
- p-value is close to 0 in both cases. Hence the coefficients of regression are statistically significant
- Adjusted R-square is 0.016 indicating that the model can explain only 1.6% of variation in total number of diabetics patient
- Again, this model was fitted on absolute numbers and we will switch to a more better approach of percentages.

Checking linearity assumptions

```python
import statsmodels.api as sm
import matplotlib.pyplot as plt
import scipy.stats as stats
import numpy as np

# Assuming you have already fitted the linear regression model named 'model'

# Get fitted values and residuals
fitted_values = stats_model.fittedvalues
residuals = stats_model.resid

# 1. Residuals vs. Fitted Values Plot
plt.scatter(fitted_values, residuals)
plt.xlabel('Fitted Values')
plt.ylabel('Residuals')
plt.title('Residuals vs. Fitted Values')
plt.axhline(y=0, color='r', linestyle='--')  # Add a horizontal line at y=0
plt.show()

# 2. Normal Q-Q Plot
sm.qqplot(residuals, line='s')
plt.title('Normal Q-Q Plot')
plt.show()

# 3. Scale-Location (Spread-Location) Plot
sqrt_standardized_residuals = np.sqrt(np.abs(stats_model.get_influence().resid_studentized_internal))
plt.scatter(fitted_values, sqrt_standardized_residuals)
plt.xlabel('Fitted Values')
plt.ylabel('Square Root of Standardized Residuals')
plt.title('Scale-Location (Spread-Location) Plot')
plt.axhline(y=0, color='r', linestyle='--')  # Add a horizontal line at y=0
plt.show()

# 4. Residuals vs. Predictor Plot for 'Grocery Tax'
plt.scatter(df_diab_age_pivot['Grocery Tax'], residuals)
plt.xlabel('Grocery Tax')
plt.ylabel('Residuals')
plt.title('Residuals vs. Grocery_Tax')
plt.axhline(y=0, color='r', linestyle='--')  # Add a horizontal line at y=0
plt.show()
```

#### Conclusions:
For the control model with age on response variable percentage diabetics, we observe that the assumptions are violated. Hence, we are applying transformation to provide remedy.

```python
#### Applying the Box-Cox transformation

# Apply Box-Cox Transformation to Response Variable
y_transformed, lambda_value = stats.boxcox(y + 1)  # Adding 1 to handle zero values

# Visualize the Transformed Response Variable
plt.hist(y_transformed, bins=20, edgecolor='black')
plt.title('Histogram of Transformed Response Variable')
plt.show()

# Check Lambda Value
print(f'Lambda Value: {lambda_value}')

# Fit a new model using the transformed response variable
stats_model_transformed = sm.OLS(y_transformed, X).fit()

# Print the summary of the transformed model
print(stats_model_transformed.summary())
```

#### Conclusions:
- The lambda value of transformation is 0.51
- For this lambda value, the model response variable (y) = % diabetics is now transformed to y^0.51 or (% diabetics)^0.51
- This changes some of the inferences.
##### New Inferences:
- A one unit increase in grocery tax increases transformed y by 4.19%
- A one unit increase in restaurant tax increases transformed y by 1.23%
- If the age bucket is 45-65, there is an increase in transformed y by 3.15%
- If the age bucket is 65-beyond, there is an increase in transformed y by 4.85%
- Adj R-square is 90.9% indicating the model explains 90.9% variability in transformed y.

#### Final Conclusion:
Overall conclusion from regression analysis:
- The diabetics % increases with increase in taxes (grocery or restaurant)
