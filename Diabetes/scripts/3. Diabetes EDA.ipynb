### Importing necessary libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm

# code to ignore warnings
import warnings
warnings.filterwarnings("ignore")

root_csv = '../csv files/'
root_pickle = '../pickle files/'
```

### Reading Files

```python
df_diab_total_pivot = pd.read_pickle(root_pickle +'Diabetes_Total_Pivot.pkl')
df_diab_total_pivot.head(5)
```

### Exploratory Data Analysis

Before moving to Scatter plot and Pearson Correlation, lets focus only on total numbers and percentage and remove duplicates established due to gender and age:

#### 1. Grocery Tax

##### a. Total Absolute Diabetics vs Grocery Tax

```python
plt.scatter(df_diab_total_pivot['Grocery Tax'], df_diab_total_pivot['Total Absolute'], marker='o', color='blue', alpha=0.7)

# Add labels and title
plt.xlabel('Grocery Tax Rate')
plt.ylabel('Total Absolute')
plt.title('Scatter Plot of Total Absolute Diabetics and Grocery Tax')
```

##### Conclusion:
At higher grocery tax rates, the total absolute number is concentrated to lower bandwidth. However, this is not a causation. It is possible that since we are looking at totals, the counties/states with high grocery tax rate have less population and hence less absolute number in diabetics. Hence we will draw a plot on Percentage

##### b. Total Percentage Diabetics vs Grocery Tax

```python
plt.scatter(df_diab_total_pivot['Grocery Tax'], df_diab_total_pivot['Total Percentage'], marker='o', color='blue', alpha=0.7)

# Add labels and title
plt.xlabel('Grocery Tax Rate')
plt.ylabel('Total Percentage')
plt.title('Scatter Plot of Total Percentage Diabetics and Grocery Tax')
```

##### Conclusion:
This plot helps us quash the possible causation based on absolute number. Looks like the percentage diabetics is not much influenced by grocery tax. Lets do a Pearson correlation test to understand more.

##### c. Pearson Correlation of Total Absolute with Grocery Tax

```python
correlation = df_diab_total_pivot['Grocery Tax'].corr(df_diab_total_pivot['Total Absolute'])
print('Grocery Tax - Total Absolute ',correlation)
```

##### Conclusion:
This correlation shows that total absolute has negative correlation with grocery tax i.e. as grocery tax increases the total absolute number decreases. However, this is again on absolute numbers and does not account for population number. Hence we will do a Pearson Correlation on Percentage.

##### d. Pearson Correlation of Total Percentage with Grocery Tax

```python
correlation = df_diab_total_pivot['Grocery Tax'].corr(df_diab_total_pivot['Total Percentage'])
print('Grocery Tax - Total Percentage',correlation)
```

##### Conclusion:
This correlation shows that total percentage has a minor positive correlation with grocery tax i.e. as grocery tax increases the total percentage of diabetics patient increases.
