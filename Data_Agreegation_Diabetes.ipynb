```python
import pandas as pd
import numpy as np
from urllib.request import urlopen
import json
import plotly.express as px
```

```python
root_csv = 'Diabetes//csv files/'
root_pickle = 'Diabetes/pickle files/'
df_diab_total_pivot = pd.read_pickle(root_pickle +'Diabetes_Total_Pivot.pkl')
df_diab_total_pivot.head(5)
```

```python
del df_diab_total_pivot['County']
del df_diab_total_pivot['CountyFIPS']
```

## 1. Taking mean of county level tax to form state level tax | Diabetes data

```python
df_diab_total_pivot_mean_grouped=pd.DataFrame()

for year in range(2004,2021):
    tdf=df_diab_total_pivot[df_diab_total_pivot['Year']==year]
    tdf=tdf.groupby('State').agg({
        'Total Absolute':'sum',
        'Total Percentage':'mean',
        'Grocery Tax':'mean',
        'Restaurant Tax':'mean',
        'tax_delta':'mean',
        'Population':'sum',
        'Year':'mean'
    })
    df_diab_total_pivot_mean_grouped = pd.concat([tdf,df_diab_total_pivot_mean_grouped])
    
```

```python
df_diab_total_pivot_mean_grouped
```

## 2. Taking median of county level tax to form state level tax | Diabetes data

```python
df_diab_total_pivot_median_grouped=pd.DataFrame()

for year in range(2004,2021):
    tdf=df_diab_total_pivot[df_diab_total_pivot['Year']==year]
    tdf=tdf.groupby('State').agg({
        'Total Absolute':'sum',
        'Total Percentage':'median',
        'Grocery Tax':'median',
        'Restaurant Tax':'median',
        'tax_delta':'median',
        'Population':'sum',
        'Year':'mean'
    })
    df_diab_total_pivot_median_grouped = pd.concat([tdf,df_diab_total_pivot_median_grouped])
    
```

```python
df_diab_total_pivot_median_grouped
```

```python
df_diab_total_pivot_mean_grouped.to_pickle(root_pickle +'Diabetes_Mean_Total_Pivot.pkl')
```

```python
df_diab_total_pivot_median_grouped.to_pickle(root_pickle +'Diabetes_Median_Total_Pivot.pkl')
```

## 5. gender diabetest

```python
root_csv = 'Diabetes//csv files/'
root_pickle = 'Diabetes/pickle files/'

df_diab_gender_pivot = pd.read_pickle(root_pickle +'Diabetes_Gender_Pivot.pkl')
```

```python
df_diab_gender_pivot.head()
```

```python
del df_diab_gender_pivot['County']
del df_diab_gender_pivot['CountyFIPS']
```

```python
df_diab_gender_pivot_mean_grouped=pd.DataFrame()

for year in range(2004,2021):
    for age in (df_diab_gender_pivot['Gender'].unique()):
        tdf=df_diab_gender_pivot[(df_diab_gender_pivot['Year']==year) & (df_diab_gender_pivot['Gender']==age)]
        tdf=tdf.groupby('State').agg({
            'Gender Absolute':'sum',
            'Gender Percentage':'mean',
            'Grocery Tax':'mean',
            'Restaurant Tax':'mean',
            'tax_delta':'mean',
            'Population':'sum',
            'Year':'mean',
            'Gender':'first'
        })
        
        tdf['tax_delta'] = tdf['Restaurant Tax'] - tdf['Grocery Tax']
        tdf['Population'] = (tdf['Gender Absolute']*100) / tdf['Gender Percentage']
    
        df_diab_gender_pivot_mean_grouped = pd.concat([tdf,df_diab_gender_pivot_mean_grouped])

   
```

```python
df_diab_gender_pivot_mean_grouped.to_pickle(root_pickle +'Diabetes_Mean_Gender_Pivot.pkl')
```

```python
from PIL import Image
import glob

def make_gif(output_path,files):
    # Create a list of image file paths
    frames=[]
    for f in files:
        frames.append(Image.open(f) )
    # Save the first frame as a GIF and append the rest
    frames[0].save(output_path, format='GIF', append_images=frames[1:], save_all=True, duration=800, loop=0)

    # Call the function with the path to your images and the desired output GIF path
# make_gif('/', 'output.gif',)
```
